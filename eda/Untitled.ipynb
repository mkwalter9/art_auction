```python
import os, sys
import pandas as pd
import numpy as np
import json

REL_DATA_PATH = '../data_collection/records'
```

```python
sys.path.append(REL_DATA_PATH)
```

```python
data_files=os.listdir(REL_DATA_PATH)
```

```python
data_files
```

```python
df = pd.DataFrame()
for n in data_files:
    df=df.append(pd.read_csv(f"{REL_DATA_PATH}/{n}",sep='|').drop('Unnamed: 0',axis=1))
```

```python
df.describe(include='all')
```

```python
df['Year Created'] = df['Year Created'].astype('int')
```

```python
for n in df.columns:
    print(f"{n} with NAN:")
    print("-----")
    print(df[n].isna().value_counts())
    print("==========================")
```
